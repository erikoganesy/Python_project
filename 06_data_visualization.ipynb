```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.options.display.float_format = '{:,.2f}'.format
```

```python
df_instacart = pd.read_parquet("data/processed/instacart.parquet")
df_instacart.head()
```

```python
orders_by_gender = (
    df_instacart[["order_id", 'Gender']]
    .drop_duplicates()
    .groupby("Gender")
    .size()
)

plt.figure()
plt.bar(orders_by_gender.index, orders_by_gender.values)
plt.title("Number of Orders by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Orders")
plt.show()
```

```python
orders_by_region = df_instacart[['order_id', 'region']].drop_duplicates().groupby('region').size()

plt.figure()
plt.bar(orders_by_region.index, orders_by_region.values)
plt.title("Number of Orders by Region")
plt.xlabel("Region")
plt.ylabel("Number of Orders")
plt.show()
```

```python
orders_by_fam_status = (
    df_instacart[['order_id', 'fam_status']]
    .drop_duplicates()
    .groupby('fam_status')
    .size()
)
plt.figure()
plt.bar(orders_by_fam_status.index, orders_by_fam_status.values)
plt.title("Number of Orders by Family Status")
plt.xlabel("Family Status")
plt.ylabel("Number of Orders")
plt.show()
```
