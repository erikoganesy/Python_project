**TASK 1** 

- Compute revenues including tax.
- Compute final revenues after tax and discount.
- Create a Boolean list indicating whether revenue > 100. 
- Create a Boolean list indicating whether revenue is between 100 and 300. 
- Add Markdown explanation: Why do parentheses matter in the formula?

```python
revenues = [120, 250, 80, 310, 95]
tax_rate = 0.18
discount_rate = 0.10
```

```python
#Revenues including tax
rev_inc_tax = [r*(1+tax_rate) for r in revenues]
print("Revenues with tax:", rev_inc_tax)
```

```python
#Final revenues after tax and discount
rev_inc_tax_disc = [r*(1+tax_rate)*(1 - discount_rate) for r in revenues]
print("Final revenues:", rev_inc_tax_disc)
```

```python
#Boolean list indicating whether revenue > 100.
lst = [i>100 for i in revenues]
print("Greater than 100:", lst)
```

```python
#Boolean list indicating whether revenue is between 100 and 300. 
between_100_300 = [i>100 and i<300 for i in revenues]
print("Between 100 and 300:", between_100_300)
```

**Add Markdown explanation: Why do parentheses matter in the formula?**

Because parentheses control the order of operations.

**TASK 2**

- Compute total revenue manually using a loop.
- Compute average revenue.
- Identify the maximum value without using max().
- Count how many values are greater than 150.
- Create a new list of revenues after adding 5% commission.

```python
sales = [120, 250, 80, 310, 95]
```

```python
#Compute total revenue manually using a loop.
total_revenue = 0
for i in sales:
    total_revenue += i
print(f"total_revenue = {total_revenue}")
```

```python
#Compute average revenue.
total_revenue = 0
for i in sales:
    total_revenue += i
avg_revenue = total_revenue/len(sales)
print(f"average_revenue = {avg_revenue}")
```

```python
#Identify the maximum value without using max()
max_r = 0
for i in sales:
    if i > max_r:
        max_r = i
print(f"max_revenue = {max_r}")
```

```python
#Count how many values are greater than 150.
count_greater_150 = sum([count > 150 for count in sales])
print(f"how many values are greater than 150 - {count_greater_150}")
```

```python
# new list of revenues after adding 5% commission
new_lst = [(i * 0.95) for i in sales]
print(f"new list of revenues after adding 5% commission - {new_lst}")
```

**TASK 3**

- “Premium” if revenue > 250
- “Standard” if 100 < revenue ≤ 250
- “Low” otherwise

1. Create a list of segment labels.
2. Count how many customers fall into each segment.
3. Add Markdown explanation: Why does the order of elif statements matter?

```python
# Create a list of segment labels
sales = [120, 250, 80, 310, 95]
segment_labels = []
for i in sales:
    if i > 250:
        segment_labels.append("Premium")
    elif i > 100:
        segment_labels.append("Standard")
    else:
        segment_labels.append("Low")
print(f"list of segment labels - {segment_labels}")

        
```

```python
#Count how many customers fall into each segment.
premium_count = segment_labels.count("Premium")
standard_count = segment_labels.count("Standard")
low_count = segment_labels.count("Low")

print("Premium:", premium_count)
print("Standard:", standard_count)
print("Low:", low_count)

```

**Why does the order of elif statements matter?**

The if–elif–else structure is checked from top to bottom.
Python stops checking conditions as soon as one condition is True.

**TASK 4**

- Remove duplicates using a set.
- Convert back to a list.
- Compare lengths before and after deduplication.
- Explain in Markdown why sets are unordered.

```python
customer_ids = [1, 2, 3, 3, 4, 5, 5, 6]
```

```python
# Remove duplicates using a set
set_customer_ids = set(customer_ids)
print(f"Removed duplicates - {set_customer_ids}")
```

```python
# Convert back to a list.
conv_lst = list(set_customer_ids)
print(f"Convert back to a list - {conv_lst}")
```

```python
# Compare lengths before and after deduplication.

len_with_dupl = len(customer_ids)
len_without_dupl = len(conv_lst)

print(f"len with duplicates - {len_with_dupl},\nlen without duplicates - {len_without_dupl}")
```

**why sets are unordered?**

Sets are implemented using a hash table.
A hash table:
- Stores elements based on their hash value
- Optimizes fast lookup (O(1) average time complexity)
- Does NOT store elements based on insertion order (in traditional set behavior)        
Because elements are placed according to their hash value in memory, their position depends on:
- The hash function
- Internal memory layout
- Python runtime behavior
Not on the order you inserted them.

**TASK 5**

- Add a new key "segment" based on revenue.
- Update revenue to 300.
- Remove "city".
- Loop over the dictionary and print key-value pairs.
- Add Markdown explanation: Why are dictionaries useful for structured data?

```python
customer = {
    "name": "Anna",
    "revenue": 250,
    "city": "Yerevan"
}
```

```python
# Add a new key "segment" based on revenue
if customer["revenue"] > 250:
    customer["segment"] = "Premium"
elif customer["revenue"] > 100:
    customer["segment"] = "Standard"
else:
    customer["segment"] = "Low"
print(f"customer with segment column- {customer}")
```

```python
# Update revenue to 300
customer["revenue"] = 300
print(f"updated customer dictionary- {customer}")
```

```python
# Remove "city" 

removed_value = customer.pop("city", None)
print(f"Removed data: {removed_value}")
print(f"Updated dictionary - {customer}")
```

```python
# Loop over the dictionary and print key-value pairs.
for key, value in customer.items():
    print(f"key, value - {key, value}")
```

**Why are dictionaries useful for structured data?**

Dictionaries are very useful for structured data because they store information in **key–value pairs**.

This makes the data:

- Organized  
- Easy to understand  
- Easy to access  
- Flexible  

**Bad corners**

Dictionaries are powerful and efficient for structured data,
but they:
- Require immutable keys
- Use more memory
- Need careful key handling
- Are not ideal for ordered or index-based tasks

3. Create a list labeling each revenue as:
    - “High” if > 200
    - “Normal” otherwise
4. Compare list comprehension vs loop in Markdown.

```python
revenues = [120, 250, 80, 310, 95]
```

```python
# Create a list of revenues after 10% tax.
revenues_after_tax = [rev_tax * 0.9 for rev_tax in revenues]
print(f"revenues after 10% tax - {revenues_after_tax}")
```

```python
# Create a list of revenues greater than 100
revenues_greater_100 = [i for i in revenues if i > 100]
print(f"revenues greater than 100 - {revenues_greater_100}")
```

```python
# Create a list labeling each revenue.
labels = []
for i in revenues:
    if i > 200:
        labels.append("High")
    else:
        labels.append("Normal")
print(f"list labeling each revenue - {labels}")
```

**Compare list comprehension vs loop in Markdown.**

| Feature     | List Comprehension     | For Loop                |
| ----------- | ---------------------- | ----------------------- |
| Code Length | Short                  | Longer                  |
| Readability | Compact                | Clear for complex logic |
| Performance | Slightly Faster        | Slightly Slower         |
| Best For    | Simple transformations | Complex logic           |
